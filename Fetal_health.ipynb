## Features and description
This dataset contains 2126 records of features extracted from the Cardiotocogram exams, which were then classified by three expert obstetritians into 3 classes: "Normal","Suspect", and "Pathological".
Dataset has the following features:

### Attributes



*  baseline value: Baseline Fetal Heart Rate (FHR) (beats per minute)
* accelerations: Number of accelerations per second
* fetal_movement: Number of fetal movements per second
* uterine_contractions: Number of uterine contractions per second
* light_decelerations: Number of light decelerations (LDs) per second
* severe_decelerations: Number of severe decelerations (SDs) per second
* prolongued_decelerations: Number of prolonged decelerations (PDs) per second
* abnormal_short_term_variability: Percentage of time with abnormal short term variability
* mean_value_of_short_term_variability: Mean value of short term variability
* percentage_of_time_with_abnormal_long_term_variability: Percentage of time * with abnormal long term variability
* histogram_width: Width of histogram made using all values from a record
* histogram_min: Histogram minimum value
* histogram_max: Histogram maximum value
* histogram_number_of_peaks: Number of peaks in the exam histogram
* histogram_number_of_zeroes: Number of zeros in the exam histogram
* histogram_mode: Histogram mode
* histogram_mean: Histogram mean
* histogram_median: Histogram median
* histogram_variance: Histogram variance
* histogram_tendency: Histogram tendency
* fetal_health: Encoded as 1-Normal; 2-Suspect; 3-Pathological

## Exploration through Visualization
Here, we use several visualization libraries/techniques to get insights from the fetal data set.

```python
# stanard import
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

We will start with the quantitive analysis to get better insights about the features.


```python
# Finding the count of fetal_health feature.
colors = ['green', 'pink', 'red']
plt.figure(figsize=(10, 10))
ax = sns.countplot(x='fetal_health', data=fetal_data, palette = colors)
ax.set_title('Fetal health distribution', fontsize=16)
plt.xticks([0, 1, 2], ['Normal', 'Suspect', 'Pathological'])
ax.set_xlabel('Fetal health status', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
```

From this countplot, we can see that the fetal health status classes are not evenly distributed. The "Normal" class has a much higher count, while the "Suspect" and "Pathological" classes are underrepresented. This imbalance can cause the model to be biased towards predicting "Normal" more often. Therefore, extra care is needed during the training process to address this issue.

Different evaluation metrics like precision,recall, F1 score and confusion matrics can be helpful in this data set.








### Multivariate Analysis
Before analyzing each feature individually, we'll start with a multivariate analysis. This gives us a bigger picture of the relationships between the features in the dataset and helps identify which features have higher correlations.

```python
# Finding the relationship among the features using correlation matrix
corr_matrix = fetal_data.corr()

# Generate a heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# let's find out the correlation of fetal health with other features.
corr_matrix = fetal_data.corr()
corr_matrix['fetal_health'].sort_values(ascending = False)
```

The correlation values show that Prolonged Decelerations (0.49) and Abnormal Short-Term Variability (0.47) have the strongest positive relationships with fetal health, indicating their importance in predicting abnormal conditions. Percentage of Time with Abnormal Long-Term Variability (0.42) also shows a significant correlation. Features like Histogram Variance (0.21) and Baseline Value (0.15) have moderate positive correlations, while others like Uterine Contractions (-0.21) and Mean Value of Long-Term Variability (-0.23) show negative correlations, meaning higher values might indicate poorer fetal health. Features with low or near-zero correlations may be less relevant for prediction. Overall, focusing on the highly correlated features will be crucial in your analysis and model development.
We would do further analysis viually on the those feature that have high correlation which can be positive and negative correlations as well.
